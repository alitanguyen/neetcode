## 1. Length of Linked List
* Iteration 
* Recursion 


```python
class Node: 
    def __init__(self, data):
        self.data = data
        self.next = None  
    
def count_nodes(head):
    # initialize count with 0 
    count = 0 

    # initialize curr with head of the linked list 
    curr = head         #curr is a pointer that points to the current node you are at  

    while curr is not None: 
        count += 1
        curr = curr.next 
    return count 

if __name__ == "__main__":
    head = Node(1)
    head.next = Node(3)
    head.next.next = Node(4)
    head.next.next.next = Node(5)

    print("Number of node is", count_nodes(head))
```

1. Why self.next = None?

* Means the node points to nothing when it's first created.
* Why None? None is used to mark the end of the list — when count_nodes hits None, it knows there are no more nodes.

    - When you first create a node: head = Node(1). It looks like this: 

        ***Node(1) -> None (points to nothing yet)***

    - Then when you link it through head.next = Node(3). It becomes: 

        ***Node(1) -> Node(3) -> None***

* Visualisation: 

    ***head -> Node(1) -> Node(3) -> Node(4) -> Node(5) -> None ***

## 2. Search an element in a linked list 


```python
class Node: 
    def __init__(self, data):
        self.data = data
        self.next = None # When a node is created alone, its .next is None by default, but it may change later when we add more nodes 

def search_key(head, key):
    #initialize curr with the head of the linked list 
    curr = head

    # iterate over the nodes
    while curr is not None: 
        if curr.data == key:     #curr.data is a node object, not a number
            return True
        curr = curr.next 
    return False

if __name__ == "__main__":

    # create a hard-coded linked list:
    # 1 -> 2 -> 3
    head = Node(1)
    head.next = Node(2)
    head.next.next = Node(3)

    # key to search in the linked list
    key = 2

    if search_key(head, key):
        print("true")
    else: 
        print("false")
```

-> curr.data = 1 == 2? No 

-> curr = curr.next -> Move to Node(2)

3.1. At the front

```python
class Node: 
    def __init__(self,data):
        self.data = data
        self.next = None 

# insert a new node at the beginning of the node
def insertAtFront(head, x):
    newNode = Node(x)               # create a new node with the value x 
    newNode.next = head             # KEY STEP 1: Your next node is whatever the current head is
    return newNode                  # KEY STEP 2? Why do we need to return? 

# print the contents of the linked list 
def printList(head):
    curr = head
    
    while curr is not None: 
        print(curr.data, end="")    # do not move to next line 
        if curr.next is not None: 
            print(" -> ", end="")
        curr = curr.next 
    print()


if __name__ == "__main__":
    # Create the linked list 2->3->4->5
    head = Node(2)
    head.next = Node(3)
    head.next.next = Node(4)
    head.next.next.next = Node(5)

    # Insert a new node at the front of the list 
    x = 8
    head = insertAtFront(head,x)

    printList(head)                 # Why not print(head)? 

```

* local variable = variable exists only inside a function. When function ends, all local variables disappear. 
* If you want something created inside the function to be used later, you usually return it and store it in another variable outside the function. In this case, it is stored in another variable named head = insertAtFront(head,x)


```python
# Example for local variable: 
def create_number():
    x = 5
    return x

a = create_number() 
print (a)
# When the function ends, x disappears but a still exists. Why? Because a is assigned to the returned value 
# With "return x" -> result = 5  
# Without "return x" -> result = None
```

3.2 At the end 

```python
class Node: 
    def __init__(self, data):
        self.data = data
        self.next = None
    
def insert_at_tail(head,x):
    newNode = Node(x)

    # edge case: if the linked list is empty 
    # make the new node as the head and return it 
    if head is None:
        return newNode
    
    # normal case
    curr = head
    while curr.next is not None: 
        curr = curr.next 
    
    # change the pointer of the last node to the new node
    curr.next = newNode

    # return the head of the list 
    return head

# print out the linked list 
def print_list(head):
    curr = head
    while curr is not None: 
        print(curr.data)
        curr = curr.next 
    

if __name__ == "__main__":
    # Create the linked list 2->3->4->5
    head = Node(2)
    head.next = Node(3)
    head.next.next = Node(4)
    head.next.next.next = Node(5)

    # Insert a new node at the end of the list 
    x = 19
    head = insert_at_tail(head,x)

    print_list(head)  
```

3.2 In the middle (after a given node)


```python
class Node: 
    def __init__(self, data):
        self.data = data
        self.next = None 
        
    @staticmethod
    def insert_middle(head, x, key):
        # create a new node
        newNode = Node(x)

        curr = head
        while curr.next is not None: 
  
            if curr.data == key: 
                newNode.next = curr.next
                curr.next = newNode
                break 

            curr = curr.next 

        return head
    
    # print out the linked list 
    @staticmethod
    def print_list(head):
        curr = head
        while curr is not None: 
            print(curr.data)
            curr = curr.next 
        

if __name__ == "__main__":
    # Create the linked list 2->3->4->5
    head = Node(2)
    head.next = Node(3)
    head.next.next = Node(4)
    head.next.next.next = Node(5)

    # Insert a new node in the middle of the list, after Node(3)
    x = 19
    key = 3
    middle = Node.insert_middle(head, x, key)

    Node.print_list(middle)  
        
 

        

        
        
```

* Why using @staticmethod? 
- No self needed like normal method 
- This is just a normal function sitting inside the box (or the class). It does not care about any of the class object. 
- When you write @staticmethod? => Don't pass self automatically
